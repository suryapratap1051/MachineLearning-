# normal distribution

```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1234)
```

```python
samples=np.random.lognormal(mean=1.,sigma=0.4,size=10000)
```

```python
shape,loc,scale=scipy.stats.lognorm.fit(samples,floc=0)
```

```python
num_bins=50
```

```python
count,edges,patches=plt.hist(samples,bins=num_bins,color='r')
```
